```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('Open_data/YouBike/YouBike場站建置成本.csv', encoding='big5')
df1
```

```python
df1['金額（土木施工費用）'] = df1['金額（土木施工費用）'].replace(',', '', regex=True)
df1
```

```python
df1['金額（土木施工費用）'] = df1['金額（土木施工費用）'].astype(int)
df1
```

```python
df11 = df1.sort_values(by='金額（土木施工費用）', ascending=False, ignore_index=True)
df11
```

```python
df2 = pd.read_csv('Open_data/YouBike/YouBike站位每月熱門站點.csv', encoding='big5')
df2
```

```python
fund = []

for i in df2['站位名稱']:
    count = 0
    for j in df1['站名']:
        if i == j:
            fund.append(df1.iloc[count, 2])
            break
        if count == len(df1)-1:
            fund.append(np.nan)
        count += 1
df2['建置成本'] = fund
df2
```
